Area plot

Mô tả xu hướng 5 quốc gia có tỷ lệ nhập cư cao nhất (1980-2013)

```python
# Chuyển các cột từ 1980 đến 2013 và các cột số khác (nếu có) sang kiểu số
df_can.iloc[:, df_can.columns.get_loc(1980):df_can.columns.get_loc(2013)+1] = df_can.iloc[:, df_can.columns.get_loc(1980):df_can.columns.get_loc(2013)+1].apply(pd.to_numeric, errors='coerce')

# Thêm cột 'Total' là tổng các cột theo trục hàng (axis=1)
df_can['Total'] = df_can.iloc[:, df_can.columns.get_loc(1980):df_can.columns.get_loc(2013)+1].sum(axis=1)
```

```python
df_can
```

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

df_top5 = df_can.head()
df_top5
```
